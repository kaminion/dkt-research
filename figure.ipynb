```python
import os 
import pickle
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

model_name = 'dkvmn+'

abs_path = f'.{os.path.sep}ckpts{os.path.sep}{model_name}{os.path.sep}'
path = f'ASSIST2009{os.path.sep}'
cnt_file_name = 'q_cnts.pkl'
acc_file_name = 'q_accs.pkl'
cnt_file = None
acc_file = None

full_path = os.path.join(os.path.join(abs_path, path), cnt_file_name)
with open(full_path, "rb") as f:
    cnt_file = pickle.load(f)
full_path = os.path.join(os.path.join(abs_path, path), acc_file_name)
with open(full_path, "rb") as f:
    acc_file = pickle.load(f)
```

```python
question_tags = []
accs = []
cnts = []

for ((cnt_k, cnt_v), (acc_k, acc_v)) in zip(cnt_file.items(), acc_file.items()):
    question_tags.append(f"{cnt_k} ({cnt_v})")
    cnts.append(cnt_v)
    accs.append(acc_v)
question_tags
```

```python
plt.rcParams['figure.figsize'] = [10, 13]

x = np.arange(len(accs))
bar_labels = plt.barh(x, accs, color="red", alpha=0.5)
plt.xlabel('Accuracy', fontsize=15, fontweight="bold")
plt.ylabel('Question Tags (count)', fontsize=15, fontweight="bold")
plt.yticks(x, question_tags, fontsize=15, rotation=0)

plt.bar_label(bar_labels)

plt.show()
```
